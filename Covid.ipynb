```python
df.drop(['Latitude','Longitude','Death','Total Confirmed cases (Indian National)',
       'Total Confirmed cases ( Foreign National )','Cured/Discharged/Migrated'],axis = 1,inplace =True)
```

```python
d_dict = {}
s_dict= {}
leng = len(df['Date'])
for i in range(0,leng):
    d_dict[str(df['Date'][i])] = 0

for i in range(0,leng):
    s_dict[str(df['Name of State / UT'][i])] = 0

print(d_dict)
```

```python
d_dict.values()
```

```python
dates = list(d_dict.keys())
states = list(s_dict.keys())
```

```python
df = df.groupby(['Name of State / UT'])
```

```python
df.first()
```

```python
s_dict.keys()
```

```python
alllst =[]
for i in states:
    fgroup = df.get_group(i)
    mdate = list(fgroup['Date'])
    tcase = list(fgroup['Total Confirmed cases'])
    lst =[]
    lst.append(int(tcase[0]))  
    for j in range(1,len(tcase)):
        if (tcase[j]-tcase[j-1])<0:
            lst.append(0)
        else:
            lst.append(tcase[j]-tcase[j-1])
    for j in range(0,len(mdate)):
        temp = []
        temp.append(mdate[j])
        temp.append(lst[j])
        alllst.append(temp)
```

```python
alllst
```

```python
df1 = pd.DataFrame(alllst,columns = ['Date','New Cases Each day'])
df1.head()
```

```python
fgroup = df1.groupby('Date')
lst = []
for i in sorted(dates):
    ngroup = fgroup.get_group(i)
    lst.append(sum(ngroup['New Cases Each day']))
```

```python
print(lst)
```

```python
N = lst
beta = [0.9,0.8,0.7,0.6,0.5]
A1 = [0]
A2 = [0]
A3 = [0]
A4 = [0]
A5 = [0]
for i in range(1,len(N)):
    A1.append(0.9*A1[i-1] + 0.1*N[i])
    A2.append(0.8*A2[i-1] + 0.2*N[i])
    A3.append(0.7*A3[i-1] + 0.3*N[i])
    A4.append(0.6*A4[i-1] + 0.4*N[i])
    A5.append(0.5*A5[i-1] + 0.5*N[i])
```

```python
for i in range(len(dates)):
    dates[i] = dates[i][0:10]
dates
```

```python
plt.figure(figsize = (14,3))
plt.plot(dates, A1)
plt.plot(dates, A2)
plt.plot(dates, A3)
plt.plot(dates, A4)
plt.plot(dates, A5)
plt.xticks(rotation='vertical',fontsize = 7)
plt.legend(["0.9", "0.8","0.7","0.6","0.5"],loc="center",prop={'size': 10})
plt.text(50.0,1500.0,'Variation of Beta')
plt.title("Moving Average Graph",fontsize = 15)
plt.show()
```
